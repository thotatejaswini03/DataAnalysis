## matplotlib-python visualization library

### What is matplotlib?

matplotlib is a python library for creating static ,animated and interactive visualizations

### Why we need it?

it helps to understand trends,patterns,and outliers

### Where we use it?

---1.dataAnalysis
---2.machine learning(for plotting the loss curves)
---3.business purposes(to analyze the market trends,finance,and scientific research)

### import matplotlib.pyplot as 

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [10,20,25,30,40]
 
plt.plot(x, y)
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
categories = ['A','B','C','D']
values = [10,15,7,12]
 
plt.bar(categories, values, color='green')
plt.title("Bar Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x = [5,7,8,7,2,17,2,9,4,11]
y = [99,86,87,88,100,86,103,87,94,78]

plt.scatter(x, y, color='red', marker='x')
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,3,5,7,11]
 
plt.plot(x, y, color='purple', linestyle='--', marker='o', label='Line 1')
plt.title("Customized Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
data = [22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]
 
plt.hist(data, bins=5, color='orange', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.plot(x, y)
plt.title("Save Plot Example")
plt.savefig("my_plot.png")  # Saves in current directory
plt.show()
```

```python
labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [215, 130, 245, 210]
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']
 
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart Example")
plt.show()
```

```python

data = [25, 30, 35, 35, 40, 45, 50, 55, 60]
plt.boxplot(data)
plt.title("Box Plot Example (Statistical Distribution)")
plt.ylabel("Values")
plt.show()
 
```

```python
import numpy as np
data = np.random.normal(0, 1, 100)
plt.violinplot(data)
plt.title("Violin Plot Example")
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 10)
y = np.cos(x)
plt.stem(x, y)
plt.title("Stem Plot Example")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,3,5,6,8]
plt.step(x, y, where='mid', color='magenta')
plt.title("Step Plot Example")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.fill_between(x, y, color='lightblue', alpha=0.6)
plt.title("Area Plot Example") #filled plot
plt.show()
```

```python
x = np.linspace(-3, 3, 30)
y = np.linspace(-3, 3, 30)
X, Y = np.meshgrid(x, y)
Z = np.sqrt(X**2 + Y**2)
 
plt.contour(X, Y, Z, cmap='plasma')
plt.title("Contour Plot Example")
plt.colorbar()
plt.show()
```

```python
plt.contourf(X, Y, Z, cmap='viridis')
plt.title("Filled Contour Plot")
plt.colorbar()
plt.show()
```

```python
theta = np.linspace(0, 2*np.pi, 100)
r = np.abs(np.sin(2*theta))
plt.polar(theta, r)
plt.title("Polar Plot Example")
plt.show()
 
```

```python
x = np.arange(1,6)
y = np.random.randint(5,15,5)
errors = np.random.rand(5)
 
plt.errorbar(x, y, yerr=errors, fmt='o', ecolor='red', capsize=5)
plt.title("Error Bar Plot Example")
plt.show()
```

```python
categories = ['A','B','C','D']
values = [10,15,7,12]
 
plt.barh(categories, values, color='orange')
plt.title("Horizontal Bar Plot")
plt.show()
 
```

```python
days = [1,2,3,4,5]
eating = [2,3,4,3,4]
sleeping = [8,7,8,6,7]
working = [8,9,7,8,8]
 
plt.stackplot(days, eating, sleeping, working, labels=['Eating','Sleeping','Working'])
plt.legend()
plt.title("Stack Plot Example")
plt.show()
```

```python
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.hexbin(x, y, gridsize=25, cmap='Blues')
plt.colorbar()
plt.title("Hexbin Plot Example")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
X = np.linspace(-5, 5, 30)
Y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
ax.plot_surface(X, Y, Z, cmap='coolwarm')
ax.set_title("3D Surface Plot")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
 
ax.scatter(x, y, z, color='red', marker='o')
ax.set_title("3D Scatter Plot")
plt.show()
```

```python
x = np.linspace(-5, 5, 30)
y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X, Y, Z, color='green')
ax.set_title("3D Wireframe Plot")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]
z = np.zeros(5)
dx = np.ones(5)
dy = np.ones(5)
dz = [5, 2, 7, 3, 6]
 
ax.bar3d(x, y, z, dx, dy, dz, color='teal')
ax.set_title("3D Bar Plot")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
X = np.linspace(-5, 5, 50)
Y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(X, Y)
Z = np.cos(np.sqrt(X**2 + Y**2))
 
surf = ax.plot_surface(X, Y, Z, cmap='coolwarm', edgecolor='none')
fig.colorbar(surf)
ax.set_title("3D Surface with Colormap")
plt.show()
 
```

```python
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = 10 * np.outer(np.cos(u), np.sin(v))
y = 10 * np.outer(np.sin(u), np.sin(v))
z = 10 * np.outer(np.ones(np.size(u)), np.cos(v))
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, color='skyblue')
ax.set_title("3D Sphere Surface")
plt.show()
 
```

```python
X = np.linspace(-5, 5, 30)
Y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, alpha=0.7, cmap='viridis')
ax.plot_wireframe(X, Y, Z, color='black', linewidth=0.5)
ax.set_title("3D Surface + Wireframe")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
ax.contour(X, Y, Z, zdir='z', offset=-2, cmap='coolwarm')
ax.contour(X, Y, Z, zdir='x', offset=-5, cmap='coolwarm')
ax.contour(X, Y, Z, zdir='y', offset=5, cmap='coolwarm')
 
ax.set_title("3D Contour Projection on Planes")
plt.show()
 
```

```python
#1.
exams = [1, 2, 3, 4, 5]
marks = [78, 85, 69, 90, 88]
plt.figure(figsize=(6,4))
plt.plot(exams, marks, marker='o', color='blue', linestyle='-')
plt.title('Student Marks Across 5 Exams')
plt.xlabel('Exam Number')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```

```python
# 3. Scatter plot of Age vs Salary
age = [22, 25, 28, 30, 35, 40, 45, 50]
salary = [25000, 28000, 35000, 40000, 48000, 55000, 60000, 65000]
plt.scatter(age, salary, color='green', marker='*')
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
# 4. Histogram of marks scored by students
marks_class = [56, 67, 45, 89, 90, 77, 68, 72, 84, 59, 61, 75, 80, 55, 70]
plt.hist(marks_class, bins=5, color='lightblue', edgecolor='black')
plt.title('Histogram of Marks')
plt.xlabel('Marks Range')
plt.ylabel('Number of Students')
plt.show()
```

```python
# 5. Customized plot with different colors, markers, and line styles
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
plt.plot(x, y1, color='red', marker='o', linestyle='--', label='Line 1')
plt.plot(x, y2, color='blue', marker='s', linestyle='-.', label='Line 2')
plt.title('Customized Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 2. Bar chart of sales of 4 products
products = ['A', 'B', 'C', 'D']
sales = [250, 400, 150, 300]
plt.bar(products, sales, color='orange')
plt.title('Sales of 4 Products')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show()
```

```python
#6.
exams = [1, 2, 3, 4, 5]
marks = [78, 85, 50, 90, 88]

products = ['A', 'B', 'C', 'D']
sales = [250, 400, 150, 300]

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.plot(exams, marks, marker='o', color='blue')
plt.title('Line Plot - Marks')
plt.xlabel('Exam')
plt.ylabel('Marks')
plt.grid(True)

plt.subplot(1,2,2)
plt.bar(products, sales, color='orange')
plt.title('Bar Plot - Sales')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```
